```python
# 대표 컬러로 통합
def color_union(x):
    if 'Black' in x:
        x = 'Black'
        return x
    elif 'Blue' in x:
        x = 'Blue'
        return x
    elif 'Red' in x:
        x = 'Red'
        return x
    elif 'Silver' in x:
        x = 'Silver'
        return x
    elif 'Brown' in x:
        x = 'Brown'
        return x
    elif 'Orange' in x:
        x = 'Orange'
        return x
    elif 'Cream' in x:
        x = 'Cream'
        return x
    elif 'Gray' in x:
        x = 'Gray'
        return x
    elif 'Calico' in x:
        x = 'Calico'
        return x
    elif 'Chocolate' in x:
        x = 'Chocolate'
        return x
    elif 'Tortie' in x:
        x = 'Tortie'
        return x
    elif 'Yellow' in x:
        x = 'Yellow'
        return x
    elif 'Liver' in x:
        x = 'Liver'
        return x
    else:
        return x
```

```python
df_pre['Color'] = df_pre['Color'].apply(color_union)
df_pre['Color'].unique()
```

# AgeuponOutcome -> DayuponOutcome Preprocessing

```python
def DayuponOutcome(age_str):
    num, unit = age_str.split(' ')
    if unit == 'day' or unit == 'days':
        return( int(num) )
    elif unit == 'week' or unit == 'weeks':
        return(int(num) * 7 )
    elif unit == 'month' or unit == 'months':
        return(int(num) * 30 )
    elif unit == 'year' or unit == 'years':
        return(int(num) * 365 )
```

```python
df_pre['AgeuponOutcome'] = df_pre['AgeuponOutcome'].fillna('0 day')
df_pre['AgeuponOutcome'] = df_pre['AgeuponOutcome'].apply(DayuponOutcome)
df_pre.rename(columns={'AgeuponOutcome':'DayuponOutcome'}, inplace = True)
```

# Breed Preprocessing

```python
def BreedMix(x):
    if '/' in  x:
        return 1
    elif 'Mix' in x:
        return 1
    else:
        return 0
```

```python
df_pre['BreedMix'] = df_pre['Breed'].apply(BreedMix)
```

```python
def Nature_Breed(x):
    if '/' in x:
        x = x[0:x.index('/')]
        return x
    elif ' Mix' in x:
        x = x[0:-4]
        return x
    else:
        return x
```

```python
df_pre['Breed'] = df_pre['Breed'].apply(Nature_Breed)
```
