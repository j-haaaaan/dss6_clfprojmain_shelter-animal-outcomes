```python
column=['Name', 'Color_pre', 'DayuponOutcome', 'BreedMix', 'Breed_pre', 'YearMonth', 'Group']

df_X = pd.DataFrame(data = df , columns = column)

X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=0 , shuffle = True)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
%%time
polysvc = SVC(kernel="poly",probability=True).fit(X_train_std, y_train)
```

```python
%%time
rbfsvc = SVC(kernel="rbf",probability=True).fit(X_train_std, y_train)
```

```python
%%time
sigmoidsvc = SVC(kernel="sigmoid",probability=True).fit(X_train_std, y_train)
```

```python

poly_probs = polysvc.predict_proba(X_test_std)
rbf_probs = rbfsvc.predict_proba(X_test_std)
sigmoid_probs = rbfsvc.predict_proba(X_test_std)
```

```python
print('train score: {}'.format(polysvc.score(X_train_std, y_train)))
print('test score: {}'.format(polysvc.score(X_test_std, y_test)))
print('log loss: {}'.format(log_loss(y_test,poly_probs)))
print()
```

```python
print('train score: {}'.format(rbfsvc.score(X_train_std, y_train)))
print('test score: {}'.format(rbfsvc.score(X_test_std, y_test)))
print('log loss: {}'.format(log_loss(y_test, rbf_probs)))
print()
```

```python
print('train score: {}'.format(sigmoidsvc.score(X_train_std, y_train)))
print('test score: {}'.format(sigmoidsvc.score(X_test_std, y_test)))
print('log loss: {}'.format(log_loss(y_test, sigmoid_probs)))
print()
```
