```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import *
from sklearn.svm import SVC
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', 30000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df= pd.read_csv('c://dss6_out/@main@/data/df_pre.csv')
```

```python
df.tail(20)
```

```python
del df['Unnamed: 0']
```

```python
df.tail()
```

```python
columns = ['Name','DateTime','AnimalType','SexuponOutcome','AgeuponOutcome','Breed','Color','Color_pre','DayuponOutcome','AboutAge',
           'BreedMix','Breed_pre','Year','Month','YearMonth','Season','Quarter','Group','Size']
```

```python
df_X = pd.DataFrame(data = df , columns = columns )
```

```python
df_y = pd.DataFrame(data = df , columns = ['OutcomeType'])
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier


forest = ExtraTreesClassifier(n_estimators=250, random_state=0)
forest.fit(df_X, df_y)

importances = forest.feature_importances_

std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

print("Feature ranking:")
for f in range(df_X.shape[1]):
    print("%d.feature %d (%f)" % (f + 1 ,indices[f], importances[indices[f]]))
    print(columns[indices[f]])
    print()

plt.title("Feature importances")
plt.bar(range(df_X.shape[1]), importances[indices], color="r", yerr=std[indices], align="center")
plt.xticks(range(df_X.shape[1]), indices)
plt.xlim([-1, df_X.shape[1]])
plt.show()
```

변수 선정 : importance 크기 순으로 겹치지 않는 변수로 선정

SexuponOutcome , DateTime , Color , DayuponOutcome , Name , Breed , Group , Size , AnimalType , BreedMix

```python
column=['SexuponOutcome','DateTime','Color','DayuponOutcome','Name','Breed','Group','Size','AnimalType','BreedMix']
```

```python
df_X = pd.DataFrame(data = df , columns = column)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=0 , shuffle = True)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
%%time
polysvc = SVC(kernel="poly",probability=True).fit(X_train_std, y_train)
```

```python
%%time
rbfsvc = SVC(kernel="rbf",probability=True).fit(X_train_std, y_train)
```

```python
%%time
sigmoidsvc = SVC(kernel="sigmoid",probability=True).fit(X_train_std, y_train)
```

```python

poly_probs = polysvc.predict_proba(X_test_std)
rbf_probs = rbfsvc.predict_proba(X_test_std)
sigmoid_probs = rbfsvc.predict_proba(X_test_std)
```

```python
print('train score: {}'.format(polysvc.score(X_train_std, y_train)))
print('test score: {}'.format(polysvc.score(X_test_std, y_test)))
print('log loss: {}'.format(log_loss(y_test,poly_probs)))
print()
```

```python
print('train score: {}'.format(rbfsvc.score(X_train_std, y_train)))
print('test score: {}'.format(rbfsvc.score(X_test_std, y_test)))
print('log loss: {}'.format(log_loss(y_test, rbf_probs)))
print()
```

```python
print('train score: {}'.format(sigmoidsvc.score(X_train_std, y_train)))
print('test score: {}'.format(sigmoidsvc.score(X_test_std, y_test)))
print('log loss: {}'.format(log_loss(y_test, sigmoid_probs)))
print()
```
